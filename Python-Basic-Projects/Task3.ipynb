```python
'''Q5. (Positional vs Keyword Arguments)
Write a function student_info(name, age) that prints the name and age.
Call this function twice:
1. Using positional arguments.
2. Using keyword arguments.
Add comments explaining the difference.
'''

def student_info(name, age):
    print(f" Name:- {name} \n Age:- {age}")

# Positional Arguments
# Values are passed according to their position.
student_info("Omkar", 18)

# Keyword Arguments
# Values are passed using parameter names, so order does not matter.
student_info(age = 20 , name ="Jay")
```

```python
'''Q7. (Recursion - Basic)
Write a recursive function countdown(n) that prints numbers from n down to 1.
Call countdown(10).
Make sure to include the base case (stopping condition).'''

def countdown(n):
    # Base case (stopping condition).
    if n == 0:
        return 
    
    print(n)
    countdown(n-1) # Recursive function

countdown(10)
```

```python
'''Q8. (Recursion - Factorial)
Write a recursive function factorial(n) that calculates and returns the factorial of a
number.
Example: factorial(5) should return 120.
Test it with input from the user and print the result.
Add a comment explaining the base case.
'''
# Function that calculates and returns the factorial of a number.
def factorial(n):
    if n == 1 or n == 0:
        return 1
    
    return n * factorial(n - 1) 

# Main program
'''Taking input. 
input() always captures data as a string.
Taking input in int() to convert it to a number.'''

n = int(input("Enter any number:-"))
print("Factorial of",n,"is:-" ,factorial(n))

```
