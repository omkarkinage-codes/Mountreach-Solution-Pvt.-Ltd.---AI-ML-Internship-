```python
'''
Q3. (Instance Methods and self)

Create a dictionary: 
Create a class BankAccount with: 
•	Attributes: account_holder and balance (initialized in __init__). 
•	Methods: deposit(amount) — add amount to balance. 
•	withdraw(amount) — subtract amount from balance (only if sufficient balance). 
•	show_balance() — print current balance. 
Create an object and demonstrate all methods.
'''


class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"\nSuccessfully deposited: Rs.{amount}")
        else:
            print("\nDeposit amount must be greater than zero!")

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"\nInsufficient balance. Available: Rs.{self.balance}")
        elif amount <= 0:
            print("\nWithdrawal amount must be greater than zero!")
        else:
            self.balance -= amount
            print(f"\nSuccessfully withdrew: Rs.{amount}")

    def show_balance(self):
        print(f"\nAccount Holder: {self.account_holder} | Current Balance: Rs.{self.balance}")

# Creating a bank account object
account = BankAccount("Omkar", 100000)
account.show_balance()

# depositing money
account.deposit(5000)
account.show_balance()

# withdrawal money
account.withdraw(30000)
account.show_balance()

# failed withdrawal
account.withdraw(200000)
account.show_balance()
    
```

```python
'''
Q4. (Method with Parameters)

Create a class Student with attributes name and marks. 
Add a method calculate_grade() that returns "Pass" if marks ≥ 40, otherwise "Fail".
Create 3 student objects, display their details, and print their grades.
'''


# Create a class Student with attributes name and marks. 
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

# Add a method calculate_grade() that returns "Pass" if marks ≥ 40, otherwise "Fail".
    def calculate_grade(self):
        if self.marks >= 40:
            return "Pass"
        else:
            return "Fail"
        
# Create 3 student objects, display their details, and print their grades.
s1 = Student("Omkar", 85)
s2 = Student("Jay", 83)
s3 = Student("Priya", 81)

print("\nGrade Calculator")
# Student 1
print(f"{s1.name:<12} | {s1.marks:<5} | {s1.calculate_grade():<5}")

# Student 2 
print(f"{s2.name:<12} | {s2.marks:<5} | {s2.calculate_grade():<5}")

# Student 3 
print(f"{s3.name:<12} | {s3.marks:<5} | {s3.calculate_grade():<5}")

```

```python
'''
Q5. (Multiple Methods)

Create a class Employee with: 
Attributes: name, salary (use __init__). 
Methods: display_details() 
give_raise(amount) — increase salary and print new salary. 
yearly_bonus() — return 10% of current salary. 
Create an employee object and demonstrate all methods.
'''

# Create a class Employee with: 
# Attributes: name, salary (use __init__).
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        
# Methods: display_details() 
    def display_details(self):
        print(f"Employee Name: {self.name} | Monthly Salary: Rs.{self.salary}")

# give_raise(amount) — increase salary and print new salary. 
    def give_raise(self, amount):
            if amount > 0:
                self.salary += amount
                print(f"Raise Approved! Added: Rs.{amount} | New Salary: Rs.{self.salary}")
            else:
                print("Raise amount must be greater than zero!")

# yearly_bonus() — return 10% of current salary. 
    def yearly_bonus(self):
            bonus_amount = self.salary * 0.10
            return bonus_amount
            

# Create an employee object and demonstrate all methods.
emp = Employee("Omkar", 50000)
print("Employee Profile")
emp.display_details()

print("\nYearly Bonus")
bonus = emp.yearly_bonus()
print(f"Calculated 10% Yearly Bonus: Rs.{bonus}")


print("\nSalary Raise")
emp.give_raise(5000)

print("\nUpdated Profile")
emp.display_details()

```

```python
'''
Q6. (Real-world Object Modeling)

Create a class MobilePhone with attributes: 
brand, model, battery_percentage. 
Add these methods: 
charge(percent) — increase battery. 
use_phone(minutes) — reduce battery by 1% per 10 minutes. 
show_status() — print current status. 
Create an object and simulate some usage.
'''


# Create a class MobilePhone with attributes: brand, model, battery_percentage. 
class MobilePhone:
    def __init__(self, brand, model, battery_percentage):
        self.brand = brand
        self.model = model
        self.battery_percentage = battery_percentage

# Add these methods: 
# charge(percent) — increase battery. 

    def charge(self, percent):
            if percent <= 0:
                print("Charging percent must be greater than zero!")
                return
                
            self.battery_percentage += percent

            if self.battery_percentage > 100:
                self.battery_percentage = 100
            print(f"Charged up by {percent}%!")

# use_phone(minutes) — reduce battery by 1% per 10 minutes. 

    def use_phone(self, minutes):
            if minutes <= 0:
                print("Usage minutes must be greater than zero!")
                return

            drain = minutes // 10
            print(f"Using the phone for {minutes} minutes... ")
            print(f"Drains {drain}% battery")

            self.battery_percentage -= drain

            if self.battery_percentage < 0:
                self.battery_percentage = 0
                print("Battery is completely dead! Please charge.")


# show_status() — print current status. 
    def show_status(self):
        print(f"Device: {self.brand} {self.model} | Battery level: {self.battery_percentage}%")


# Create an object and simulate some usage.
my_phone = MobilePhone("Google", "Pixel 9", 50)
print("\nMobilePhone Status")
my_phone.show_status()

print("\nMobilePhone Usage")
my_phone.use_phone(90)
my_phone.show_status()

print("\nMobilePhone Charging")
my_phone.charge(45)
my_phone.show_status()

print("\nMobilePhone Overcharge Protection")
my_phone.charge(20)
my_phone.show_status()
```

```python
'''
Q7. (Combining Concepts)

Create a class Rectangle with attributes length and width. 
Add methods: 
area() — return the area. 
perimeter() — return the perimeter. 
display() — print length, width, area, and perimeter. 
Take input from user for length and width, create object, and display results
'''


# Create a class Rectangle with attributes length and width. 

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

# Add methods: 
# area() — return the area. 

    def area(self):
        # Formula: Length * Width
        return self.length * self.width
    

# perimeter() — return the perimeter. 

    def perimeter(self):
        return 2 * (self.length + self.width)

# display() — print length, width, area, and perimeter. 

    def display(self):
        print("\nRectangle Dimensions & Calculations")
        print(f"Length:    {self.length}")
        print(f"Width:     {self.width}")
        print(f"Area:      {self.area()}")
        print(f"Perimeter: {self.perimeter()}")

# Take input from user for length and width, create object, and display results
try:
    print("\nEnter the dimensions for your Rectangle:")
    user_length = float(input("Enter Length: "))
    user_width = float(input("Enter Width: "))

    if user_length <= 0 or user_width <= 0:
        print("\nLength and width must be positive numbers greater than zero!")
    else:
        # Create the object using the user's 
        user_rectangle = Rectangle(user_length, user_width)
        
        # Display the results
        user_rectangle.display()

except ValueError:
    print("\nInvalid Input!")
```

```python
'''
Q8. (Updating Attributes)

Create a class Player with attributes name, score, and level. 
Add methods: 
increase_score(points) 
level_up() — increase level by 1.
show_progress() 
Create an object, update score and level multiple times, and show progress
'''


# Create a class Player with attributes name, score, and level. 
class Player:
    def __init__(self, name, score=0, level=1):
        self.name = name
        self.score = score
        self.level = level

# Add methods: 
# increase_score(points) 
    def increase_score(self, points):
            if points > 0:
                self.score += points
                print(f"\nPoints Added: +{points} XP!")
            else:
                print("\nPoints to increase must be greater than zero!")

# level_up() — increase level by 1.

    def level_up(self):
            self.level += 1
            print("\nLEVEL UP! You reached Level" ,self.level)

# show_progress() 

    def show_progress(self):
            print(f"\nPlayer: {self.name} | Level: {self.level} | Total Score: {self.score}")


# Create an object,  

gamer = Player("Omkar")
print("Character Created")
gamer.show_progress()

# update score and level multiple times, and show progress
gamer.increase_score(150)
gamer.level_up()
gamer.show_progress()


gamer.increase_score(50)
gamer.increase_score(75)
gamer.show_progress()

```

```python
'''
Q10. (Mini Project - Combined Concepts)

Create a Simple Library Management System using OOP: 
Design a class Book with: 
Attributes: title, author, status ("Available" or "Issued"). 

Methods: 
•	issue_book(), 
•	return_book(), 
•	show_info(). 

Create multiple book objects and demonstrate issuing/returning books with proper messages. 

Use a menu (while loop) to allow the user to: 
1. Add a new book 
2. Issue a book 
3. Return a book 
4. Show all books
'''

# Creating an empty dictionary to store all book objects
library_db = {}

# Design a class Book with: 
# Attributes: title, author, status ("Available" or "Issued"). 

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.status = "Available"

    # Method issue_book
    def issue_book(self):
        if self.status == "Available":
            self.status = "Issued"
            print(f"'{self.title}' has been issued.")
        else:
            print(f"'{self.title}' is already issued to someone else!")

    # Method return_book
    def return_book(self):
        if self.status == "Issued":
            self.status = "Available"
            print(f"'{self.title}' has been returned safely.")
        else:
            print(f"'{self.title}' was never issued! It is already here.")

    # Method show_info to display book info 
    def show_info(self):
        print(f"{self.title:<25} | {self.author:<20} | {self.status:<10}")


print("\nWelcome to Simple Library Management System")

# Use a menu (while loop) to allow the user to: 
# 1. Add a new book 
# 2. Issue a book 
# 3. Return a book 
# 4. Show all books

while True:
    print("\nLibrary Menu:")
    print("1. Add a New Book")
    print("2. Issue a Book")
    print("3. Return a Book")
    print("4. Show All Books")
    print("5. Exit")
    
    # Get user choice 
    choice = input("\nEnter your choice (1-5): ").strip()

# 1. Add a new book 

    if choice == "1":
        print("\nAdd New Book")
        title = input("Enter Book Title: ").upper()
        author = input("Enter Book Author: ").upper()
        

        if title in library_db:
            print("This book already exists in the library!")
        elif title == "" or author == "":
            print("Title or Author cannot be empty!")
        else:
            library_db[title] = Book(title, author)
            print(f"'{title}' added to the library database.")

# 2. Issue a book 

    elif choice == "2":
        print("\nIssue a Book")
        title = input("Enter Book Title to issue: ").upper()
        
        # Check if the book exists inside the database
        if title in library_db:
            # call its method
            library_db[title].issue_book()
        else:
            print("Book not found in library!")

    # 3. Return a book 

    elif choice == "3":
        print("\nReturn a Book")
        title = input("Enter Book Title to return: ").upper()

    # Check if the book exists inside the database
        if title in library_db:
            # call its method
            library_db[title].return_book()
        else:
            print("Book not found in library!")

    # 4. Show all books

    elif choice == "4":
        print("\nLibrary Catalog ")
        # Check if dictionary is empty
        if not library_db:
            print("The library is completely empty right now.")
        else:
            print(f"{'Title':<25} | {'Author':<20} | {'Status':<10}")
            print("-" * 62)
            for book_obj in library_db.values():
                book_obj.show_info()

    elif choice == "5":
        print("\nThank you for using the library system. Goodbye!")
        break
        
    else:
        print("Invalid Choice! Please type a number from 1 to 5.")
```
