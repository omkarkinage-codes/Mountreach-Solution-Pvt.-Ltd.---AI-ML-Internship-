```python
'''
Q1. (Creating Dictionaries)

Write a program to create dictionaries using different methods: 
a) An empty dictionary using two different ways. 
b) A dictionary with string keys (name, city, course). 
c) A dictionary with integer keys. 
d) A mixed data type dictionary (string, int, float). 
Print each dictionary and its type. 
Add comments explaining the syntax
'''

# a) An empty dictionary using two different ways. 
print("\na) An empty dictionary using two different ways:-")
emp_dic1 = {}
print("Empty dictionary using {}:- " ,emp_dic1)
print("Type:" ,type(emp_dic1))
emp_dic2 = dict()
print("Empty dictionary using dict() method/constructor:- " ,emp_dic2)
print("Type:" ,type(emp_dic2))


# b) A dictionary with string keys (name, city, course). 
str_dict = {
    "name" : "Omkar",
    "city" : "Pune",
    "course" : "AIML"
}
print("\nb) A dictionary with string keys:-\n" ,str_dict)
print("Type:" ,type(str_dict))


# c) A dictionary with integer keys. 
int_dict = {
    1: "One",
    2: "Two",
    3: "Three",
    4: "Four" 
}
print("\nc) A dictionary with integer keys:-\n" ,int_dict)
print("Type:" ,type(int_dict))


# d) A mixed data type dictionary (string, int, float). 
mix_dict = {
    "String" : "Omkar",
    "Int" : 21,
    "Float" : 74.7
}
print("\nd) A mixed data type dictionary:-\n" ,mix_dict)
print("Type:" ,type(mix_dict))
```

```python
'''
Q2. (Accessing and Modifying Elements)

Create the dictionary: 
student = { "name": "YourName", "age": 00, "city": "YourCity", "marks": 00 } 
Print the value of "name" using square brackets. 
Update the "marks" to 92. 
Add a new key "grade" with value "A". 
Print the updated dictionary.
'''

# Give dictionary
student = { 
    "name": "Omkar", 
    "age": 20,
    "city": "Pune", 
    "marks": 98

} 
print("\nGive Dictionary:-\n" ,student)

# Printing the value of "name" using square brackets. 
print("\nPrint the value of 'name':-" ,student["name"])

# Updating the "marks" to 92. 
student.update({"marks" : 92})

# Adding a new key "grade" with value "A". 
student["grade"] = "A"

# Printing the updated dictionary.
print("\nThe updated dictionary:-\n" ,student)
```

```python
'''
Q5. (update() and pop())

Create a dictionary: 
info = {"brand": "Samsung", "model": "A52", "price": 25000} 
Update it with new information: {"color": "Black", "price": 24000}. 
Remove the key "model" using pop() and print the removed value. 
Print the final dictionary.
'''

# Give dictionary
info = {"brand": "Samsung", "model": "A52", "price": 25000} 
print("\nGive Dictionary:-\n" ,info)

# Updating it with new information: {"color": "Black", "price": 24000}. 
info.update({"color": "Black", "price": 24000})
print("\nAfter updating dictionary with new information:\n" ,info)

# Removing the key "model" using pop() and print the removed value. 
print("\nAfter Removing the key 'model' using pop();-\n" ,info.pop("model"))

# Printing the final dictionary.
print("\nThe final dictionary:-\n" ,info)
```

```python
'''
Q6. (popitem() and clear())

Create a dictionary with at least 5 key-value pairs.
Use popitem() twice and print the removed items each time. 
Clear the entire dictionary using clear(). 
Print the final result. 
Add comments explaining the difference between pop() and popitem().
'''


# Creating a dictionary with at least 5 key-value pairs.
device = {
    "brand": "Samsung",
    "model": "A52",
    "price": 25000,
    "color": "Black",
    "storage": "128GB"
}
print("\nCreated dictionary:\n" ,device)

# Using popitem() twice and print the removed items each time. 
print("\nPopped Item 1:-" ,device.popitem())
print("\nPopped Item 2:-" ,device.popitem())

# Clear the entire dictionary using clear(). 
device.clear()

# Printing the final result. 
print("\nFinal result;-" ,device)


'''
Explaining the difference between pop() and popitem().

1. dict.pop(key):
   - Targets a SPECIFIC item based on the key name you provide.
   - Returns only the VALUE of that removed key.
   - Throws a KeyError if the key you specified does not exist.

2. dict.popitem():
   - Does not take any arguments. It automatically targets the VERY LAST item currently 
     sitting at the end of the dictionary.
   - Returns the entire item as a (key, value) tuple package.
   '''
```

```python
'''
Q7. (copy() and setdefault())

Create a dictionary: d = {"a": 1, "b": 2} 
Make a copy of it.
Use setdefault() to add key "c" with value 3 (if it doesn't exist). 
Use setdefault() again for an existing key "a". 
Print the original and copied dictionaries.
'''

# Give dictionary
d = {"a": 1, "b": 2} 
print("\nGive Dictionary:-\n" ,d)

# Making a copy of it.
c = d.copy()

# Using setdefault() to add key "c" with value 3 (if it doesn't exist). 
c.setdefault("c", 3)
print("\nAdding new key using setdefault():-\n" , c)

# Using setdefault() again for an existing key "a". 
c.setdefault("a", 7)
print("\nUsing setdefault() for an existing key 'a':-\n" , c)


# Printing the original and copied dictionaries.
print("\nThe original Dictionary:-\n" ,d)
print("\nThe Copied Dictionary:-\n" ,c)
```

```python
'''
Q8. (fromkeys() and Membership)

Write a program that: 
Uses dict.fromkeys() to create a dictionary with keys ["name", "age", "city"] and default value None. 
Takes user input to fill these keys. 
Checks if a given key exists using in operator.
'''


d = dict.fromkeys(["name", "age", "city"], None)
print("\nDictionary:-\n" ,d)

# Taking user input to fill these keys. 
print("\nEnter Details")
for key in d:
    d[key] = input(f"\nEnter value for '{key}':- ").lower()

print("\nUpdated Dictionary:-\n", d)


# Checking if a given key exists using in operator.
user_inp = input("\nEnter a key name to check if it exists:- ")
if user_inp in d:
    print(f"\nYes! '{user_inp}' is present in the Dictionary")
else:
    print(f"\nNo! '{user_inp}' is Not present in the Dictionary")
```

```python
'''
Q9. (Practical Application)

Write a program to store and manage contact information using a dictionary: 
Take name, phone number, and email as input. 
Store them in a dictionary with name as key. 
Allow the user to search a contact by name using get(). 
Print all contacts using items().
'''

print("\nWellcome to Contacts Manager")

contacts = {}

num = int(input("How many contacts do you want to add? :- "))

print("\nEnter Contact Details ")
# Taking name, phone number, and email as input. 
for i in range(num): 
    print(f"\nAdding Contact {i+1}:")
    name = input("Enter Name:- ").lower()
    phone = input("Enter Phone Number:- ").lower()
    email = input("Enter Email Address:- ").lower()
    contacts[name] = {"phone" : phone, "email" : email}



print("\nData recorded successfully.")

# Allow the user to search a contact by name using get(). 
print("\nSearch Contact ")
name_inp = input("Enter a name to search for:- ").lower()

result = contacts.get(name_inp)
if result:
    print(f"\nContact Found!")
    print(f"Phone: {result['phone']}")
    print(f"Email: {result['email']}")
else:
    print("\nContact not found.")


# Printing all contacts using items().
print("\nDisplaying All Contacts")
# 'name' and 'info' are created right here on this line!
for name, info in contacts.items():
    print(f"Name: {name} | Phone: {info['phone']} | Email: {info['email']}")
```

```python
'''
Q10. (Mini Project - Combined Concepts)

Create a Student Management System using a dictionary:
Repeatedly ask the user to enter student details (roll number as key, and a nested dictionary containing name, age, and marks). 
Use a while loop with a menu: 
1. Add new student 
2. Update marks of a student 
3. Search student by roll number 
4. Display all students 
5. Remove a student 
6. Exit
'''

student_db = {}

print("\nWelcome to the Student Management System ")

while True:
    print("\nSYSTEM MENU")
    print("1. Add new student")
    print("2. Update marks of a student")
    print("3. Search student by roll number")
    print("4. Display all students")
    print("5. Remove a student")
    print("6. Exit")

    choice = input("Enter your choice (1-6):- ").lower()

    if choice == "1":
        print("\nAdd New Student")
        roll_no = input("Enter Roll Number:- ").lower()
        
        # Checking if the roll number already exists 
        if roll_no in student_db:
            print(f"Roll Number '{roll_no}' already exists! Use Option 2 to update.")
        else:
            name = input("Enter Student Name:- ").lower()
            age = int(input("Enter Student Age:- "))
            marks = float(input("Enter Student Marks:- "))
            
            # Creating and storing the nested dictionary profile
            student_db[roll_no] = {
                "name": name,
                "age": age,
                "marks": marks
            }
            print(f"Student '{name}' successfully registered.")

    elif choice == "2":
        print("\nUpdate Student Marks")
        roll_no = input("Enter Roll Number to update:- ").lower()
        
        if roll_no in student_db:
            new_marks = float(input(f"Enter new marks for {student_db[roll_no]['name']}:- "))
            # Accessing the inner dictionary key directly to update the value
            student_db[roll_no]["marks"] = new_marks
            print("Student marks updated successfully.")
        else:
            print("Student roll number not found.")

    elif choice == "3":
        print("\nSearch Student Database")
        roll_no = input("Enter Roll Number to search:- ").lower()
        
        # Using .get() to safely grab the profile without throwing a KeyError crash
        student = student_db.get(roll_no)
        
        if student:
            print(f"\nStudent Profile Found:")
            print(f"Roll Number : {roll_no}")
            print(f"Name        : {student['name']}")
            print(f"Age         : {student['age']}")
            print(f"Marks       : {student['marks']}")
        else:
            print("Student Profile Not Found.")

    elif choice == "4":
        print("\nDisplaying All Registered Students")
        if not student_db:
            print("The database is completely empty! No records found.")
        else:
            for roll_no, info in student_db.items():
                print(f"Roll No: {roll_no} | Name: {info['name']} | Age: {info['age']} | Marks: {info['marks']}")

    elif choice == "5":
        print("\nRemove Student Profile")
        roll_no = input("Enter Roll Number to remove:- ").lower()
        
        if roll_no in student_db:
            # Using pop() deletes the key and returns the profile data package
            removed_student = student_db.pop(roll_no)
            print(f"Removed student '{removed_student['name']}' from the database.")
        else:
            print("Roll number does not exist.")

    elif choice == "6":
        print("\nThank you for using the Student Management System. Goodbye!")
        break  
        
    else:
        print("\nInvalid input! Please choose a valid operation from 1 to 6.")
```
