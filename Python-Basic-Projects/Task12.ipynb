```python
'''
Q2. (Strings + Loops + Functions)

Write a function analyze_string(s) that takes a string as input and:
Prints its length using len().
Prints the string in reverse using slicing.
Counts and prints how many vowels (a,e,i,o,u) are in the string (case
insensitive).
Uses a for loop with range() to print each character with its positive and
negative index.
Call this function with user input and handle empty string case.
'''

def analyze_string(s):

    if not s:
        print("The String is empty.")

    else:
        # Prints its length using len().
        print("Length of the string:- ", len(s))

        # Prints the string in reverse using slicing.
        print("String in reverse:- ", s[::-1])

        # Counts and prints how many vowels (a,e,i,o,u) are in the string
        vowels = "aeiou"
        count = 0
        for i in s:
            if s.lower() in vowels:
                count += 1
            
        print("\nThe count of vowels(a,e,i,o,u) in your String is:- ",count)

        # Uses a for loop with range() to print each character with its positive and negative index.
        print("\nCharacter with its positive and negative index")
        for i in range(len(s)):
            p_indx = i
            n_indx = i - len(s)
            print(f"Character:- '{s[i]}'  |  Postive Index:- {p_indx}  |  Negative Index:- {n_indx}"  )




print("String Operations")
user_str = input("\nEnter a string:- ")
analyze_string(user_str)



```

```python
'''
Q3. (Lists + Functions + Exception Handling)
Create a function manage_marks() that:
Takes 5 subject marks as input from the user.
Stores them in a list.
Handles ValueError if non-numeric input is given.
Calculates and prints the average, highest, and lowest marks.
Sorts the list in descending order and prints it.
Use meaningful comments.
'''

marks = []

# Create a function manage_marks() 
def manage_marks():
    
    try:
        # Takes 5 subject marks as input from the user.
        print("\nEnter marks for 5 subjects")
        for i in range(1,6):
            mark = float(input(f"Subject {i}:- "))
            if mark < 0 or mark > 100:
                print("Marks must be between 0 and 100. Try again.")
            else:
                # Stores them in a list.
                marks.append(mark)
   
        
        # Calculates and prints the average, highest, and lowest marks.
        average = sum(marks)/5
        highest = max(marks)
        lowest = min(marks)
        print("\nAverage marks = ",average)
        print("Highest marks = ",highest)
        print("Lowest marks = ",lowest)

        # Sorts the list in descending order and prints it.
        marks.sort(reverse=True)
        print("The list in descending order:-\n" ,marks)

     
    # Handles ValueError if non-numeric input is given.
    except ValueError:
        print("Invalid Input! Please enter a valid number.")


   
# Calling function
manage_marks()
```

```python
'''
Q4. (OOP + Lists + Exception Handling)

Create a class Student with attributes: 
name, roll_no, and marks_list (a list of marks). 
Include methods: __init__ to initialize the student. 
add_mark(mark) to add a mark to the list (handle invalid marks). 
get_average() to return the average. 
display_info() to show all details. 
Create one student object, add marks using user input, and demonstrate all methods with exception handling.
'''


class Student:
    # __init__ to initialize the student.
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no
        self.marks_list = []

# add_mark(mark) to add a mark to the list . 
    def add_mark(self, mark):
        if mark < 0 or mark > 100:
            raise ValueError("Marks must be between 0 and 100.")
        self.marks_list.append(mark)
    
# get_average() to return the average. 
    def get_average(self):
        if not self.marks_list:
            return 0.0
        return sum(self.marks_list) / len(self.marks_list)
    
    # display_info() to show all details. 
    def display_info(self):
        print("\nSTUDENT PROFILE ")
        print(f"Name         : {self.name}")
        print(f"Roll Number  : {self.roll_no}")
        print(f"Marks List   : {self.marks_list}")
        print(f"Average Mark : {self.get_average():.2f}")

# Taking input from user
print("\nRegister New Student ")
name_input = input("Enter Student Name: ").upper()
roll_input = input("Enter Roll Number: ")

# Creating object 
student_obj = Student(name_input, roll_input)

# Ask the user how many marks they wish to add
try:
    num_of_marks = int(input("\nHow many marks do you want to add?: "))
except ValueError:
    print("Invalid count! Defaulting to 3 marks entries.")
    num_of_marks = 3

# Collect marks via input loop with exception handling
i = 1
while i <= num_of_marks:
    try:
        user_mark = float(input(f"Enter mark {i}: "))

        # Call the object method 
        student_obj.add_mark(user_mark)
        i += 1  

    except ValueError as error:
        print(f"{error} Please re-enter a valid score.")

#  Calling display_info()
student_obj.display_info()
```

```python
'''
Q5. (Dictionaries + Functions + Control Structures)
Write a function student_database() that uses a dictionary (roll number as key) to
store student records.
Provide a menu using while loop:
1. Add student (name, age, city)
2. Search student by roll number
3. Display all students
4. Exit
Use get(), update(), and proper exception handling for invalid inputs.
'''

def student_database():
    database = {}

    while True:
        #  Menu Options
        print("\nStudent Database System ")
        print("1. Add Student")
        print("2. Search Student by Roll Number")
        print("3. Display All Students")
        print("4. Exit")

        choice = input("\nEnter your choice (1-4): ")

        # Exit Condition
        if choice == "4":
            print("\nExiting Database. Goodbye!")
            break

        # Add a Student 
        elif choice == "1":
            try:
                roll_no = input("\nEnter Roll Number (Key): ")
                if not roll_no:
                    print("Roll number cannot be empty.")
                    continue

                # Check if roll number already exists 
                if database.get(roll_no) is not None:
                    print(f"Student with Roll Number '{roll_no}' already exists!")
                    continue

                name = input("Enter Name: ").upper()
                age = int(input("Enter Age: "))  

                if age < 0:
                    print("Age cannot be negative.")
                    continue

                city = input("Enter City: ").upper()

                # Store details in a inner dictionary
                student_details = {"name": name, "age": age, "city": city}

                database.update({roll_no: student_details})
                print(f"Student record added for Roll No {roll_no}!")

            except ValueError:
                print("Invalid input! Age must be a whole number.")

        # Search Student by Roll Number
        elif choice == "2":
            search_roll = input("\nEnter Roll Number to search: ")

            student = database.get(search_roll)

            if student:
                print(f"\nRecord Found for Roll No {search_roll} ")
                print(f"Name : {student['name']}")
                print(f"Age  : {student['age']}")
                print(f"City : {student['city']}")
            else:
                print(f"No student found with Roll Number '{search_roll}'.")

        # Display All Students
        elif choice == "3":
            if not database:
                print("\nThe database is currently empty.")
            else:
                print("\nAll Student Records ")
                for roll, info in database.items():
                    print(f"Roll No: {roll} | Name: {info['name']} | Age: {info['age']} | City: {info['city']}")

        # Handle wrong menu choices
        else:
            print("Invalid Choice! Please select a valid number from 1 to 4.")


student_database()
```

```python
'''
Q6. (Sets + Tuples + Modules)
Write a program that:
Takes 10 numbers as input and stores unique numbers in a set.
Converts that set into a tuple.
Uses random module to pick and print 3 random numbers from the tuple.
Uses math module to print the square root of the sum of the tuple elements.
Handle possible exceptions.
'''

import random
import math

try:
    nums = set()

# Takes 10 numbers as input and stores unique numbers in a set.
    for i in range(1,11):
        num = float(input(f"\nEnter number {i}:- "))
        nums.add(num)
    print("\nSet of Numbers :-\n" ,nums)

# Converts that set into a tuple.
    print("\nConverted the Set into Tuple:-")
    num_tup = tuple(nums)
    print(num_tup)

# Uses random module to pick and print 3 random numbers from the tuple.
    print("\nPrinting Three random numbers from Set:-")
    if len(num_tup) >= 3:
        for i in range(1,4):
            r_num = random.choice(num_tup)
            print(f"Random nuber {i}:- ",r_num)
    else:
        print("Not enough unique numbers entered to select 3 random values.")

# Uses math module to print the square root of the sum of the tuple elements.
    tup_sum = sum(num_tup)
    print("\nSum of all tuple elements:- ",tup_sum)
    sum_sqr = math.sqrt(tup_sum)
    print("The square root of the sum of the tuple elements:-" ,sum_sqr)


except ValueError:
    print("Invalid Input! Please Enter a number.")
```

```python
'''
Q7. (Lambda + range() + Lists + Exception Handling)
Create a lambda function to calculate the square of a number.
Write a program that:
Uses range(1, 21) to generate numbers.
Stores the squares of these numbers in a list using the lambda function.
Prints only the even squares from the list.
Add exception handling if needed.
'''

try:
    #Create a lambda function to calculate the square of a number
    square = lambda x: x**2

# Stores the squares of these numbers in a list using the lambda function.
    sq_list = []
    for num in range(1, 21):
        sq_list.append(square(num))

    print("All Squares (1 to 20):")
    print(sq_list)

    #  Print only the even squares from the list
    print("\nEven Squares from the list:")
    even_sq = []
    for sq in sq_list:
        if sq % 2 == 0:
            even_sq.append(sq)

    print(even_sq)

# Add exception handling if needed.
except Exception as error:
    print("An unexpected error occurred")
```

```python
'''Q8. (Tuples + Dictionaries + OOP)

Create a class Employee with: 
Attributes: emp_id, name, details (a tuple containing department and salary).
Method show_details() to print all information. 
Create a dictionary where employee ID is the key and Employee object is the value. 
Add 3 employees and display all using a loop.
'''

class Employee:
    def __init__(self, emp_id, name, department,salary):
        self.emp_id = emp_id
        self.name = name
        self.details = (department,salary)


    def show_details(self):
        department, salary = self.details
        print(f"ID: {self.emp_id:<5} | Name: {self.name:<10} | Dept: {department:<25} | Salary: ₹{salary}")

    
# Create a dictionary where employee ID is the key and Employee object is the value. 
employee_directory = {}

# Add 3 employees and display all using a loop.
emp1 = Employee(101, "Omkar", "AI & ML Engineering", 75000)
emp2 = Employee(102, "Jay", "Data Science", 68000)
emp3 = Employee(103, "Priya", "Software Development", 72000)

employee_directory[emp1.emp_id] = emp1
employee_directory[emp2.emp_id] = emp2
employee_directory[emp3.emp_id] = emp3

print("\nAll Employee Records ")
for emp_object in employee_directory.values():
    emp_object.show_details()
```

```python
'''
Q9. (Strings + Sets + Exception Handling + Modules)
Write a program that:
Takes a sentence as input.
Extracts all unique words (using set) from the sentence.
Prints the unique words in sorted order.
Uses math module to print the total number of unique words raised to power 2.
Handle any possible exceptions gracefully.
'''

import math

try:
    sentence = input("\nEnter a sentence:- ").lower()
    if not sentence:
        raise ValueError("The sentence cannot be empty.")
    
    # Extracts all unique words (using set) from the sentence.
    words_list = sentence
    unique_words = set(words_list)


    # Prints the unique words in sorted order.
    print("\nUnique words in sorted order:")
    print(sorted(unique_words))

    # Uses math module to print the total number of unique words raised to power 2.
    total_count = len(unique_words)
    print("\nTotal number of unique words:-" ,total_count)
    print("Unique words count raised to the power of 2:-",math.pow(total_count,2))

except ValueError as val_error:
    print(f"Input Error: {val_error}")
except Exception as unexpected_error:
    print(f"An unexpected error occurred: {unexpected_error}")
```

```python
'''
Q10. (Mini Project - Comprehensive)

Create a Smart Calculator & Data Manager program that combines multiple concepts: 
Use a while loop menu with these options: 
1. Basic Arithmetic (use functions + exception handling) 
2. Scientific Calculations (use math module)
3. Generate Random Numbers (use random module) 
4. Store Results in Dictionary (with timestamp using string) 
5. View History (show stored results) 
6. Exit

'''

import datetime
import math
import random


def basic_arithmetic(num1, num2, operation):
    if operation == "+":
        return num1 + num2
    elif operation == "-":
        return num1 - num2
    elif operation == "*":
        return num1 * num2
    elif operation == "/":
        if num2 == 0:
            raise ZeroDivisionError("Division by zero is undefined.")
        return num1 / num2


def smart_calculator():
    history_log = {}

    while True:
        print("\nSMART CALCULATOR & DATA MANAGER ")
        print("1. Basic Arithmetic")
        print("2. Scientific Calculations (Power / Square Root)")
        print("3. Generate Random Numbers")
        print("4. View Calculation History")
        print("5. Exit")

        choice = input("Select an option (1-5): ")

        if choice == "5":
            print("\nShutting down system. Thank you for using Smart Calculator!")
            break

        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        try:
            # Basic Math Operations
            if choice == "1":
                n1 = float(input("Enter first number: "))
                n2 = float(input("Enter second number: "))
                op = input("Enter operator (+, -, *, /): ")

                if op not in ["+", "-", "*", "/"]:
                    print("Invalid mathematical operator selected.")
                    continue

                result = basic_arithmetic(n1, n2, op)
                summary_text = (f"Arithmetic: {n1} {op} {n2} = {result}")
                print(f"\nResult: {result}")

                # Save operational metrics into history
                history_log[timestamp] = summary_text

            # Scientific Math Routines
            elif choice == "2":
                print("\nScientific Options ")
                print("a. Square Root (sqrt)")
                print("b. Power (x raised to y)")
                sub_choice = input("Choose operation (a/b): ").lower()

                if sub_choice == "a":
                    val = float(input("Enter number: "))
                    if val < 0:
                        raise ValueError("Square root of a negative value is invalid.")
                    
                    result = math.sqrt(val)
                    summary_text = (f"Scientific: sqrt({val}) = {result}")

                elif sub_choice == "b":
                    base = float(input("Enter base (x): "))
                    exponent = float(input("Enter power (y): "))
                    result = math.pow(base, exponent)
                    summary_text = (f"Scientific: {base} ** {exponent} = {result}")

                else:
                    print("Invalid sub-choice selection.")
                    continue

                print(f"\nResult: {result}")
                history_log[timestamp] = summary_text

            # Random Range Computations
            elif choice == "3":
                lower_bound = int(input("Enter minimum integer boundary: "))
                upper_bound = int(input("Enter maximum integer boundary: "))

                if lower_bound > upper_bound:
                    raise ValueError("Logic Error: Lower bound cannot be higher than upper bound.")

                result = random.randint(lower_bound, upper_bound)
                summary_text = (f"Random: Integer between [{lower_bound}, {upper_bound}] generated -> {result}")

                print(f"\nGenerated Random Value: {result}")
                history_log[timestamp] = summary_text

            # Database Retrieval Logic
            elif choice == "4":
                print("\n CALCULATION LOG HISTORY ")
                if not history_log:
                    print("No logs recorded yet.")
                else:
                    for time_key, log_detail in history_log.items():
                        print(f"[{time_key}] {log_detail}")
                

            else:
                print("Out of bounds menu choice! Please select 1 through 5.")

        except ValueError as val_err:
            print(f"\n[Execution Failure] Input Type/Domain Error: {val_err}")
        except ZeroDivisionError as zero_err:
            print(f"\n[Execution Failure] {zero_err}")
        except Exception as general_err:
            print(f"\n[Execution Failure] Unexpected system collision: {general_err}")


if __name__ == "__main__":
    smart_calculator()
```
