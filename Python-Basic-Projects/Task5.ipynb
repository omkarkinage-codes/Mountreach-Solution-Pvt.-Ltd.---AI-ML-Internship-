```python
'''
Q1. (String Indexing) 

Write a program that takes a string input from the user. 
Print: 
•	The first character 
•	The last character (using negative index)
•	The 3rd character from the start 
•	The 2nd character from the end 
Add comments explaining positive and negative indexing.
'''


# Taking string input from the user (input() is a string by default)
x = input("Enter a String:-")
print("Your entered String is:-" ,x)

# Positive indexing: Starts from 0 from the left side
print("\nThe 1st character of the string is:-" ,x[0])

# Negative indexing: Starts from -1 from the right side
print("\nThe last character of the string is:-" ,x[-1])

# Positive index 2 gives the 3rd character
print("\nThe 3rd character of the string is:-" ,x[2])

# Negative index -2 gives the second character from the end
print("\nThe last 2nd character of the string is:-" ,x[-2])


'''
Positive Indexing: Goes left-to-right, starting at 0. 
Negative Indexing: Goes right-to-left, starting at -1.

Example for string "OMKAR":
 O   M   K   A   R  
 0   1   2   3   4   = Positive
-5  -4  -3  -2  -1   = Negative
'''
```

```python
'''
Q2. (String Slicing)

Take a string input from the user.
Using slicing, print: 
a) The first 5 characters 
b) The last 4 characters 
c) The string in reverse order 
d) Every 2nd character of the string 
Include comments explaining the slicing syntax [start:end:step].
'''

# SLICING SYNTAX :
# Syntax: string[start : end : step]


while True:
    
# Taking string input from the user (input() is a string by default)
    x = input("Enter a String:-")
    print("Your entered String is:-" ,x)

    # Checking that the string has enough characters to process
    if len(x) <= 5 :
        print("\nEntered string is too short!")
        print("Please Enter new string...\n")

    else:
        # a) The first 5 characters 
        print("\nThe first 5 characters of string are:-" ,x[:5])
        # b) The last 4 characters 
        print("\nThe last 4 characters of string are:-" ,x[-4:])
        # c) The string in reverse order
        print("\nThe string in reverse order:-" ,x[: : -1])
        # d) Every 2nd character of the string 
        print("\nThe Every 2nd character of the string :-" ,x[: : 2])
        
        break
    

```

```python
'''Q4. (len() Function) 

Write a program that: 
Takes a string as input. 
Prints the length of the string using len(). 
Prints the last character using len() (without using negative index). 
Prints the middle character if the length is odd. 
Add comments on common mistakes with len().
'''


# Taking string input from the user (input() is a string by default)
s = input("Enter main string:-")
print("Your entered main string is:-" ,s)

# 1. Printing the total length of the string using len()
print("\nThe lenght of entered string is:-" ,len(s))

# 2. Printing the last character using len()
print("\nThe last character of string is:-", s[len(s) - 1])

# 3. Checking if the length is odd
if len(s) % 2 != 0:
    print("\nThe middle character of string is:-" ,s[len(s) // 2])
else:
    print("\nThe string length is even, " \
    "so there is no single middle character.")

"""
COMMON MISTAKES WITH len():
1. The Off-by-One Error (IndexError):
   Trying to access s[len(s-1)] to get the last character. Because indexing starts 
   at 0, the last valid index is always len(s) - 1. s[len(s-1)] will crash the program.

2. Using len() on numbers:
   Passing an integer or a float directly into the function (e.g., len(125)). 
   len() only works on sequences like strings, lists, or tuples.
"""
```

```python
'''Q5. (range() - Basic Forms)

Write a program using for loop and range() to:
 a) Print numbers from 0 to 10.
 b) Print numbers from 5 to 15. 
 c) Print odd numbers from 1 to 21. 
Explain the three forms of range() in comments.
'''

print("a) Print numbers from 0 to 10.")
# Form 1: range(stop) -> Starts at 0 by default
for i in range(11):
    print(i)

print("\nb) Print numbers from 5 to 15.")
# Form 2: range(start, stop) -> Starts at 5, stops before 16 (counts by 1)
for i in range(5,16):
    print(i)

print("\nc) Print odd numbers from 1 to 21.")
# Form 3: range(start, stop, step) -> Starts at 1, stops before 22, skips by 2
for i in range(1,22,2):
    print(i)
```

```python
'''Q6. (range() with Negative Step)

Use range() to print: 
Numbers from 20 down to 10 (decreasing). 
Numbers from 15 down to 1 in steps of 2. 
Print each sequence on a new line with proper messages.
'''
'''
NEGATIVE STEP RANGE :
Syntax: range(start, stop, negative_step)
'''

# 1. Print numbers from 20 down to 10
# We start at 20, stop at 9, and step of -1
print("1. Printing numbers from 20 down to 10")
for i in range(20,9,-1):
    print(i)

# 2. Print numbers from 15 down to 1 in steps of 2
# We start at 15, stop at 0, and step of -2
print("\n2. Printig numbers from 15 down to 1 in steps of 2")
for i in range(15,0,-2):
    print(i)
```

```python
'''Q7. (Combined: Strings + range())

Write a program that takes a string from the user. 
Using a for loop and range() with len(), print: 
Each character of the string one by one (with its index). 
The string in reverse order using negative step in range().
'''


# Taking string input from the user (input() is a string by default)
x = input("Enter a String:-")
print("Your entered String is:-" ,x)

for i in range(len(x)):
    print(f"Character of string:- {x[i]}  is on Index: {i}")


print("\nPrinting the string in reverse order:-")
# 2. Loop backward using range(start, stop, step)
# Start: len(x) - 1 (index of the very last character)
# Stop: -1 (stops right before index -1, meaning it includes index 0)
# Step: -1 (tells the loop to count backward)
for i in range(len(x) -1,-1,-1):
    print(x[i])
```

```python
'''Q9. (Slicing + range())

Write a program that prints the following using range() and string slicing concepts where applicable: 
First 10 even numbers (2 to 20). 
Numbers from 1 to 30 in steps of 3. 
A string "PythonProgramming" sliced to show "Python", "Programming", and the reverse.
'''

print("First 10 even numbers (2 to 20). ")
for i in range(2,21,2):
    print(i)


print("\nNumbers from 1 to 30 in steps of 3. ")
for i in range(1,31,3):
    print(i)


s = "PythonProgramming"
print("\nOriginal String:-", s)

# Slicing "Python" 
print("\nSliced 'Python':-", s[:6])

# Slicing "Programming" 
print("Sliced 'Programming':-", s[6:])

# Reversing the full string using step -1
print("Reversed String:-", s[::-1])
```

```python
'''Q10. (Mini Project - Combined) 

Create a String Analyzer program: 
Take a string input from the user. 
Using functions or directly: 
1. Print length of the string. 
2. Print first half and second half using slicing. 
3. Check if the word "python" (case insensitive — you may convert) is present. 
4. Use range() to print all characters with their positive and negative indices. 
5. Print the string in reverse. 
Add proper comments and use meaningful variable names.
'''

# Taking string input from the user (input() is a string by default)
user_str = input("Enter a String:-")
print("Your entered String is:-" ,user_str)

# 1. Print length of the string. 
print("\nThe length of string is:-" ,len(user_str))


# 2. Print first half and second half using slicing. 
half_str = len(user_str) // 2
print("\nFirst half of the string:-" ,user_str[:half_str])
print("Second half of the string:-" ,user_str[half_str:])


# 3. Check if the word "python" (case insensitive — you may convert) is present. 
lowercase_str = user_str.lower()
if "python" in lowercase_str:
    print("\nThe word 'python' IS present in your string!")
else:
    print("\nThe word 'python' is NOT present in your string.\n")


# 4. Use range() to print all characters with their positive and negative indices. 
for i in range(len(user_str)):
    neg_index = i - len(user_str)

    print(f"Character: '{user_str[i]}' | Positive Index: {i} | Negative Index: {neg_index}")


# 5. Print the string in reverse. 
print("\nReverse String:-" ,user_str[::-1])
```
