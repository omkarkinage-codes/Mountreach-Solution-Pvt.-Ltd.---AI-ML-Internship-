```python
'''Q1. (Lambda Function)
Write a lambda function named square that takes a number x and returns its
square.
Call the lambda function with input from the user and print the result.
Also write the same logic using a normal def function and compare both in
comments.
'''
'''
Taking input. 
input() always captures data as a string.
Taking input in int() to convert it to a number.'''

num = int(input("Enter any nmber:-"))

# Lambda function

sqr = lambda x: x**2
# Calling the Lambda function
print("\nBy Lambda function")
print(f"The square of {num} is:-" ,sqr(num))


# Normal def function 

def square(x):
    return x**2
    
# Calling the normal function
print("\nBy Normal def function")
print(f"The square of {num} is:-" ,square(num))

"""
COMPARISON:
1. Lambda is a 1-line function with no name or 'return' keyword. Normal functions use 'def' and 'return'.
2. Lambda is for quick, simple math. Normal functions are better for long, multi-step code.
"""
```

```python
'''
Q2. (Lambda with Multiple Arguments) 
Create a lambda function add_three(a, b, c) that returns the sum of three numbers.
Take three numbers as input from the user and print their sum using this lambda function.
'''

'''
Taking input. 
input() always captures data as a string.
Taking input in int() to convert it to a number.'''

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

# Lambda function
sum = lambda a,b,c: a+b+c

# Calling the normal function
print("\nThe sum of entered numbers is:-" ,sum(a,b,c))
```

```python
'''
Q3. (For Loop Revision)
Write a program using a for loop and range() to:
a) Print all numbers from 1 to 25.
b) Print the sum of all numbers from 1 to 20.
c) Print the table of 5 (5 × 1 to 5 × 10).
Add comments explaining when to use for loop'''

# Variable to hold the total sum
sum = 0

# LOOP 1: Handles printing 1-25 and calculating the sum up to 20
for i in range (1, 26):
    if i == 1:
        print("Printing all numbers from 1-25.")
    print(i)

    if i <= 20:
        sum+=i
    
print("\nPrinting the sum of all numbers from 1 to 20.")
print("The sum of all numbers from 1 to 20 is:-" ,sum)


# LOOP 2: to print the multiplication table
print("\nPrinting the table of 5 (5 × 1 to 5 × 10).")
for i in range (1,11):
    print(f"5 * {i} = {5*i}")
```

```python
'''
Q4. (While Loop Revision)
Write a program using a while loop that:
Repeatedly asks the user to enter a positive number.
Keeps adding the numbers to a running total.
Stops when the user enters 0 or a negative number.
Finally prints the total sum.
'''

# Variable to hold the total sum
sum= 0 

# Start an infinite loop that keeps running until we break it
while True:
    '''
Taking input. 
input() always captures data as a string.
Taking input in int() to convert it to a number.'''

    num = int(input("Enter any number(0 or negative to exit):-"))

    if num <= 0:    # Check if the number is 0 or negative to stop the loop
        print("Loop terminated.")
        break
    else:
        sum += num
        
# Printing the final total sum 
print("The sum of entered numbers is:-" ,sum)
```

```python
'''
Q5. (Math Module)
Write a program that imports the math module and does the following:
Takes a number as input.
Prints its square root using math.sqrt().
Prints its factorial using math.factorial() (for positive integers only).
Prints the ceiling and floor value using math.ceil() and math.floor().
Add comments about the usefulness of the math module
'''

# Importing the built-in math module
import math


while True:
    # Taking input as a float to allow decimal numbers
    num = float(input("Enter any number:-"))


    # Check if the number is valid (0 or positive)
    if num >= 0:
        # Calculate and print the square root
        print(f"\nSquare root of {num} is:-" ,math.sqrt(num))

        # Convert to integer and calculate the factorial
        a = int(num)
        print(f"\nFactorial of {a} is:-" ,math.factorial(a))

        # Calculate and print ceiling and floor values
        print(f"\nCeiling value of {num} is:-" ,math.ceil(num))
        print(f"\nFloor value of {num} is:-" ,math.floor(num))

        break

    else:
        # Error message if the user enters a negative number
        print("Invalid Input...!")
        print("Please enter a number greater than or equal to 0.\n")
        




```

```python
'''
Q6. (Random Module)

Write a program using the random module that: 
Generates and prints 5 random integers between 1 and 100. 
Generates and prints one random number between 50 and 150. 
Prints a random floating point number between 0 and 1. 
Use different functions from the random module.
'''


import random

for i in range(1,6):
    # Function 1: Using randint(a, b)
    print(f"Random Integer {i}:-" ,random.randint(1, 100))


# Function 2: Using randrange(start, stop)
print("\nRandom Number(50-150):-" , random.randrange(50, 151))


# Function 3: Using random()
print("\nFloat Number:-" , random.random())

```

```python
'''
Q7. (Import Methods)

Demonstrate different ways of importing modules: 
1. Import the entire math module using import math and use math.pow(2, 4). 
2. Import only sqrt from math using from math import sqrt. 
3. Import math with an alias as m and use m.factorial(5).
Print the results with clear messages.
'''


# Method 1: Import the entire module
import math
print("Power of the 2 raised to 2 is:-" ,math.pow(2, 4))


# Method 2: Import only a specific function
from math import sqrt
print("\nSqure root of number:-" ,sqrt(4874))


# Method 3: Import the module with a short nickname (alias)
import math as m
print("\nFactorial of 5 is:-" ,m.factorial(5))
```
