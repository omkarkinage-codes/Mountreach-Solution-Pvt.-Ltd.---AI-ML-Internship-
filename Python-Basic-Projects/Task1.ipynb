```python
#Q5. (Input/Output)

# Taking string input from the user
name = input("Enter your name: ")

'''Taking birth year input. 
    input() always captures data as a string.
    Taking input in int() to convert it to a number.'''

birth_year = int(input("Enter your birth year: "))

current_year = 2026
cal_age = current_year - birth_year

print(f"Hello {name}! your age is:-", cal_age)
```

```python
#Q6. (Combined - Variables, Operators & Input)


print("Body Mass Index Calculator")

# input() always captures data as a string.
# Taking input in float() to convert it to a decimal number.
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

# Calculating BMI 
bmi = weight / (height ** 2)

print("Your calculated BMI is:", round(bmi, 2))
```

```python
#Q7. (Comments)

"""
This program takes the length and width of a rectangle from the user.
Then it calculates AREA and PERIMETER
Finally it prints both results on the screen.
"""

# Need to use float() here because length can be a decimal number.
length = float(input("Enter the length of the rectangle: "))

# same for width
width = float(input("Enter the width of the rectangle: "))

# Formula for area 
area = length * width

# For perimeter
perimeter = 2 * (length + width)


# Displaying the final answers to the user
print("The calculated Area of the rectangle is:", area)
print("The calculated Perimeter of the rectangle is:", perimeter)
```

```python
#Q8. (Operators - Assignment & Membership)


fruits = ["mango", "apple", "banana", "cherry"]
score = 50


# 1. Adding 25 to the score using the += shorthand operator
score += 25
print("1. The updated score after adding 25 is:", score)

# 2. Checking if "apple" is inside our fruits list
has_apple = "apple" in fruits
print("\n2. Is 'apple' in the fruits list?:", has_apple)

# 3. Checking if "grape" is missing from our fruits list
no_grape = "grape" not in fruits
print("\n3. Is 'grape' NOT in the fruits list?:", no_grape)
```

```python
#Q9. (Mini Project - All Concepts)


print("Enter Student Details")

student_name = input("Enter student name: ")
age = int(input("Enter age: "))
city = input("Enter city: ")

# Collecting marks for 3 separate subjects
subject1 = float(input("Enter marks for Subject 1: "))
subject2 = float(input("Enter marks for Subject 2: "))
subject3 = float(input("Enter marks for Subject 3: "))

# Doing the calculations
total_marks = subject1 + subject2 + subject3

# Assuming each subject is out of 100 marks (so max marks = 300)
percentage = (total_marks / 300) * 100


# Displaying the result
print("Name       :", student_name)
print("Age        :", age, "years old")
print("City       :", city)
print("\nSubject 1  :", subject1)
print("Subject 2  :", subject2)
print("Subject 3  :", subject3)
print("\nTotal Marks:", total_marks, "/ 300")

# Rounding the percentage to 2 decimal places 
print("Percentage :", round(percentage, 2), "%")
```
