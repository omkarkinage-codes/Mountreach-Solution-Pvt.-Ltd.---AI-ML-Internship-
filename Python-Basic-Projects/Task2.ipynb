```python
'''Q1. (if Statement)
Write a program that asks the user to enter their age.
If the age is 18 or more, print the message: "You are eligible to vote."
(Do not print anything if the condition is False).
'''


'''Taking input. 
    input() always captures data as a string.
    Taking input in int() to convert it to a number.'''

age = int(input("Enter your age:-"))

# Cheacking the condition
if age >= 18:
    print("You are eligible to vote.") # Executes only if condition is TRUE.
```

```python
'''Q2. (if-else Statement)
Write a program that takes a number as input from the user and checks
whether it is even or odd.
Print "Even Number" or "Odd Number" accordingly.'''


'''Taking input. 
    input() always captures data as a string.
    Taking input in int() to convert it to a number.'''

num = int(input("Enter a number:-"))

#Checking that entered number is EVEN or ODD.

if num % 2 == 0 :
    print("Entered number is EVEN.")
else:
    print("Entered number is ODD.")

```

```python
'''Q3. (if-elif-else Statement)
Write a program to assign grades based on marks entered by the user:
Marks ≥ 90 → "Grade A - Excellent"
Marks ≥ 75 → "Grade B - Good"
Marks ≥ 60 → "Grade C - Average"
Marks ≥ 40 → "Grade D - Pass"
Marks < 40 → "Fail"
Use if-elif-else and print the appropriate grade message.
'''

'''Taking input. 
    input() always captures data as a string.
    Taking input in int() to convert it to a number.'''

marks = int(input("Enter your marks:-"))

if marks < 0 or marks > 100 :
    print("Invalid marks...!")
    marks1 = int(input("Enter valid marks(0-100):-"))
    marks = marks1

print("\nYour entered marks are:-" ,marks)
    
if marks >= 90:
    print("Grade A - Excellent")
elif marks >= 75:
    print("Grade B - Good")
elif marks >= 60:
    print("Grade C - Average")
elif marks >= 40:
    print("Grade D - Pass")
else:
    print("Fail")
```

```python

'''Q4. (For Loop with range)
Write a program using for loop and range() to:
a) Print all numbers from 1 to 30.
b) Print all odd numbers from 1 to 50.
c) Print the sum of numbers from 1 to 15.
'''

# a) Print all numbers from 1 to 30.
print("All numbers from 1 to 30.") 
# range(start, stop)
# start = 1, stop = 31
for i in range (1,31) :
    print(i)

    
# b) Print all odd numbers from 1 to 50.
print("\nAll ODD numbers from 1 to 50.")
# range(start, stop, step) 
# sart = 1, stop = 51, step = 2 - to skip every 2nd number
for i in range (1,51, 2) :
    print(i)


#c) Print the sum of numbers from 1 to 15.
print("\nSum of numbers from 1 to 15.")
sum = 0
for i in range (1,16) :
    sum += i
print(sum)
```

```python
'''Q6. (While Loop - User Controlled)
Write a program that repeatedly asks the user to enter a positive number.
Keep adding the entered numbers to a running total.
Stop when the user enters a negative number or zero.
Finally, print the total sum.'''


print("Enter positive numbers to add them up. \nEnter 0 or a negative number to stop.")

sum = 0
# first input from the user 
num = float(input("\nEnter a number: "))

while num > 0:
    sum += num
    
    # Asking user for the NEXT number inside the loop
    num = float(input(f"Enter another number: "))

print("\nLoop stopped because a zero or negative number was entered.")
print("\nThe final total sum of all positive numbers is:", sum)
```

```python
'''Q7. (Nested if Statement)
Write a program that takes two inputs:
temperature (integer)
is_raining (enter "yes" or "no")
Then give advice using nested if:
If temperature > 30 and not raining → "Hot day, wear light clothes."
If temperature > 30 and raining → "Hot and rainy, carry umbrella."
If temperature < 15 and raining → "Cold and rainy, wear jacket and take
umbrella."
Else give a suitable message.
'''

temp = int(input("Enter temperature:-"))
is_raining = input("Is it raining? (Enter 'yes' or ''no'):-")

# Checking for temperature > 30
if temp > 30:
    if is_raining == "no":
        print("Advice: Hot day, wear light clothes.")
    elif is_raining == "yes":
        print("Advice: Hot and rainy, carry umbrella.")
    else:
        print("Advice: It is hot! (Unrecognized rain input).")

#  Checking for temperature < 15
elif temp < 15:
    if is_raining == "yes":
        print("Advice: Cold and rainy, wear jacket and take umbrella.")
    else:
        print("Advice: It's cold outside, make sure to stay warm!")

# Suitable message for anything else 
else:
    print("Advice: The weather is moderate and pleasant. Have a great day!")


```

```python
'''Q8. (For Loop + if-elif-else)
Write a program using for loop that goes from 1 to 40 and:
Prints "Fizz" if number is divisible by 3
Prints "Buzz" if number is divisible by 5
Prints "FizzBuzz" if divisible by both 3 and 5
Otherwise prints the number
'''

for i in range(1,41) :
    if i % 3 == 0 and i % 5 == 0 : 
        print("FizzBuzz")
    elif i % 3 == 0 :
        print("Fizz")
    elif i % 5 == 0 : 
        print("Buzz")
    else:
        print(i)
        
```
