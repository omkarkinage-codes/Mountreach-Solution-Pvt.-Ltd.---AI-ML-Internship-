```python
'''
Q1. (Basic try-except)

Write a program that takes two numbers as input from the user and prints their division. 
Use try-except to handle the case when the user enters non-numeric input (ValueError). 
Print a friendly message if an error occurs.
'''


try:
    num1 = float(input("\nEnter 1st number:-"))
    num2 = float(input("Enter 2nd number:-"))
    print("\nAnswer is:- " ,num1/num2)
except ValueError:
    print("Invalid Input! Please Enter a number...")


```

```python
'''
Q2. (Handling ZeroDivisionError)

Write a program that divides two numbers entered by the user.
Use try-except to specifically catch ZeroDivisionError and print:
"Error: Division by zero is not allowed."
Also handle ValueError for invalid input.
'''

try:
    num1 = float(input("\nEnter 1st number:-"))
    num2 = float(input("Enter 2nd number:-"))
    print("\nAnswer is:- " ,num1/num2)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Invalid Input! Please Enter a number...")
```

```python
'''
Q4. (Using else)

Write a program that takes two numbers and divides them.
Use try, except, and else. 
The else block should print "Division performed successfully!" only when no exception occurs.
'''

# Write a program that takes two numbers and divides them.
# Use try,except, and else.
try:
    num1 = float(input("\nEnter 1st number:-"))
    num2 = float(input("Enter 2nd number:-"))
    print("\nAnswer is:- " ,num1/num2)
    
except ZeroDivisionError:
    print("\nError: Division by zero is not allowed.")

except ValueError:
    print("\nError: Invalid input! You must enter numbers only.")

else:
    print("\nDivision performed successfully!")
```

```python
'''
Q5. (Using finally)

Modify the division program to include a finally block that always prints: 
"Program execution completed." 
Demonstrate that finally runs whether an exception occurs or not. 
Add comments explaining the purpose of finally.
'''

try:
    num1 = float(input("\nEnter 1st number:-"))
    num2 = float(input("Enter 2nd number:-"))
    print("\nAnswer is:- " ,num1/num2)
    
except ZeroDivisionError:
    print("\nError: Division by zero is not allowed.")

except ValueError:
    print("\nError: Invalid input! You must enter numbers only.")

# The finally block always runs, whether an error happens or not.
# It is used for cleanup tasks like closing files or connections.
finally:
    print("\nProgram execution completed." )
```

```python
'''
Q6. (Combined try-except-else-finally)

Write a complete division program using all four blocks (try, except, else, finally):
Handle ValueError and ZeroDivisionError. 
Print success message in else. 
Always thank the user in finally.
'''

print("\nDivision Program")
try:
    num1 = float(input("\nEnter 1st number:-"))
    num2 = float(input("Enter 2nd number:-"))
    print("\nAnswer is:- " ,num1/num2)
    
except ZeroDivisionError:
    print("\nError: Division by zero is not allowed.")

except ValueError:
    print("\nError: Invalid input! You must enter numbers only.")

else:
    print("\nDivision performed successfully!")

finally:
    print("\nThank you! Program execution completed." )
```

```python
'''
Q7. (Practical - Age Input)
	
Write a program that asks the user for their age. 
•	Convert the input to integer. 
•	If the age is negative, raise a ValueError with message "Age cannot be negative."
•	Handle the exception gracefully and print a suitable message.
'''

try:
    # Write a program that asks the user for their age. 
    # Convert the input to integer. 

    age = int(input("Enter your age: "))

# •	If the age is negative, raise a ValueError with message "Age cannot be negative."
    if age < 0:
        raise ValueError("Age cannot be negative.")

    print("Your age is registered as:- ",age)

# •	Handle the exception gracefully and print a suitable message.
except ValueError as error:
    print(f"\nError: {error}")
```

```python
'''
Q8. (Multiple Exceptions in One Block) 

Create a program that: 
Takes a number as input. 
Divides 100 by that number. 
Handle both ValueError and ZeroDivisionError in a single except block using a tuple. 
Print a combined error message.
'''

try:
# Create a program that: 
# Takes a number as input.
    user_input = float(input("\nEnter a number to divide 100 by: "))

# Divides 100 by that number. 
    result = 100 / user_input
    print(f"Result: 100 / {user_input} =" ,result)

# Handle both ValueError and ZeroDivisionError in a single except block using a tuple. 
# Print a combined error message.
except (ValueError, ZeroDivisionError):
    print("\nInvalid operation! Please enter a non-zero number.\n")
```

```python
'''
Q10. (Mini Project - Exception Handling)

Create a Simple Calculator with exception handling: 
The program should run in a loop and show this menu: 
text 
1. Addition 
2. Subtraction 
3. Multiplication 
4. Division 
5. Exit 

Take user choice and two numbers. 
Perform the operation inside try block. 
Handle ValueError (invalid input) and ZeroDivisionError (for division). 
Use finally to print "Operation attempted." after every calculation. 
Continue until user chooses to exit.
'''

# Create a Simple Calculator with exception handling: 
# The program should run in a loop and show this menu: 
# text 
# 1. Addition 
# 2. Subtraction 
# 3. Multiplication 
# 4. Division 
# 5. Exit

while True:  #Continue until user chooses to exit.

    # Display Menu
    print("\nSimple Calculator ")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")

# Take user choice and two numbers. 
    choice = input("Enter choice (1-5): ").strip()

    if choice == "5":
        print("Exiting Calculator. Goodbye!")
        break

    if choice not in ["1", "2", "3", "4"]:
        print("Invalid choice! Please select a number from 1 to 5.")
        continue

# Perform the operation inside try block. 
    try:
        num1 = float(input("Enter 1st number: "))
        num2 = float(input("Enter 2nd number: "))

        if choice == "1":
            print(f"Result: {num1} + {num2} = ",num1 + num2)
        elif choice == "2":
            print(f"Result: {num1} - {num2} = ",num1 - num2)
        elif choice == "3":
            print(f"Result: {num1} * {num2} = ",num1 * num2)
        elif choice == "4":
            print(f"Result: {num1} / {num2} = ",num1 / num2)

# Handle ValueError (invalid input) and ZeroDivisionError (for division). 
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")

    except ValueError:
        print("Invalid input! Please enter valid numbers.")

# Use finally to print "Operation attempted." after every calculation. 
    finally:
        print("Operation attempted.")
```
