```python
'''
Q3. (Membership Testing)

Write a program that: 
Takes a set of 5 colors as input from the user. 
Asks the user to enter a color name. 
Checks whether the color is present in the set using in and not in. 
Prints appropriate messages.
'''

# Defining an empyt set that holds user input
colors = set()

# Taking a set of 5 colors as input from the user. 
print("\nEnter color name to create a set of colors:- ")
for i in range (5):
    col = input(f"Enter color name {i+1}:-").upper()
    colors.add(col)

# Printing the set created by user
print("\nYour colors set is:-\n" ,colors)

# Asks the user to enter a color name to search
user_color = input("\nEnter a color name to check:- ").upper()

print("\nTesting with 'in'")
if user_color in colors:
    print(f"Yes! '{user_color}' is present in your color set.")
else:
    print(f"No! '{user_color}' is NOT present in your color set.")

print("\nTesting with 'not in'")
if user_color not in colors:
    print(f"Yes! '{user_color}' is NOT present in your color set.")
else:
    print(f"No! '{user_color}' IS present in your color set.")
```

```python
'''
Q4. (add(), remove(), discard())

Create a set: fruits = {'apple', 'banana', 'mango'} 
Perform the following operations and print the set after each step: 
a) Add 'orange' using add().
b) Add 'banana' again (observe the result).
c) Remove 'mango' using remove(). 
d) Try to remove 'grape' using discard() (no error should occur).

'''

# Given Set
fruits = {'apple', 'banana', 'mango'} 
print("\nGiven Set:-\n" ,fruits)

# a) Add 'orange' using add().
fruits.add("orange")
print("\nAfter adding 'orange':-\n" ,fruits)

# b) Add 'banana' again
# 'banana' already exists, so the set safely ignores this operation!
fruits.add("banana")
print("\nAfter adding 'banana':-\n" ,fruits)

# c) Remove an item using remove()
# .remove() deletes the item, but crashes with a KeyError if the item doesn't exist
fruits.remove("mango")
print("\nAfter removing 'mango':-\n", fruits)

# d) Remove an item using discard()
# .discard() deletes the item, but stays completely safe if the item is missing!
fruits.discard("grape") 
print("\nAfter discarding missing 'grape':-\n", fruits)
```

```python
'''
Q5. (pop() and clear())

Create a set: s = {100, 200, 300, 400, 500}
Use pop() to remove and print one element. 
Print the set after popping. 
Clear the entire set using clear(). 
Print the final empty set. 
Add comments explaining the difference between remove(), discard(), and pop().
'''

# Given Set
s = {100, 200, 300, 400, 500}
print("\nGiven Set:-\n" ,s)

# Use pop() to remove and print one element. 
print("\nPopped Item:-" ,s.pop())

# Print the set after popping
print("\nSet after popping:-", s)


# Clear the entire set using clear(). 
s.clear()
print("\nAfter clearing set:-", s)


'''
Difference between remove(), discard(), and pop()

1. remove(item): Removes a SPECIFIC element. 
   If the item is not in the set, it crashes with a KeyError!

2. discard(item): Also removes a SPECIFIC element. 
   If the item is missing, it does nothing and stays completely safe.

3. pop(): Takes NO arguments. It removes and returns a RANDOM element.
   If the set is empty, it crashes with a KeyError.
'''
```

```python
'''
Q6. (update() and copy())

Create two sets: set1 = {1, 2, 3} set2 = {3, 4, 5, 6} 
Make a copy of set1. 
Update set1 with elements from set2 using update(). 
Print both the original copy and the updated set.
'''

# Given Set
set1 = {1, 2, 3} 
print("\n# Given Set1:-" ,set1)
set2 = {3, 4, 5, 6} 
print("\n# Given Set2:-" ,set2)

# Make a copy of set1. 
s1 = set1.copy()

# Update set1 with elements from set2 using update(). 
set1.update(set2)

# Print both the original copy and the updated set.
print("\nAfter update() Operation ")
print("Original Copy:-", s1)
print("Updated set1:- ", set1)
```

```python
'''
Q7. (Union and Intersection)
Create two sets:
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
Print the union of A and B (using both .union() and operator).
Print the intersection of A and B (using both .intersection() and & operator).
Add comments explaining what each operation does.
'''

# Given Set
A = {1, 2, 3, 4, 5}
print("\n# Given Set1:-" ,A)

B = {4, 5, 6, 7, 8}
print("\n# Given Set2:-" ,B)

# Print the union of A and B (using both .union() and operator).
print("\nThe union of A and B (using .union()):-\n" ,A.union(B))
print("The union of A and B (using | operator):-\n" ,A | B)

# Print the intersection of A and B (using both .intersection() and & operator).
print("\nThe intersection of A and B (using .intersection()):-\n" ,A.intersection(B))
print("The intersection of A and B (using & operator):-\n" ,A & B)


'''
1. UNION (A | B): 
   Combines all unique elements from both sets. It acts like a logical "OR".
   Elements 4 and 5 are present in both, but they only appear once in the output.

2. INTERSECTION (A & B): 
   Extracts ONLY the common items that exist in both sets simultaneously. 
   It acts like a logical "AND". (Only 4 and 5 match here).
'''
```

```python
'''
Q8. (Combined Methods)

Write a program that: 
Takes 6 numbers as input from the user and stores them in a set (duplicates should be removed automatically).
Adds two more numbers using add(). 
Removes one number safely using discard(). 
Prints the final set and its length using len().
'''


# Defining an empty set that holds user input
num = set()

# Takes 6 numbers as input from the user and stores them in a set (duplicates should be removed automatically).
print("\nEnter 6 numbers to create a set of integers:- ")
for i in range (6):
    n = int(input(f"Enter number {i+1}:-"))
    num.add(n)

# Printing the set created by user
print("\nYour numbers set is:-\n" ,num)


# Adds two more numbers using add(). 
print("\nAdding Two More Numbers")
extra1 = int(input("Enter an extra number to add:- "))
num.add(extra1)

extra2 = int(input("Enter another extra number to add:- "))
num.add(extra2)

print("Set after adding new numbers:-", num)

# Removes one number safely using discard(). 
print("\nSafely Removing a Number")
r_num = int(input("Enter a number to remove from the set:- "))
# discard() is safe because it won't crash if the number isn't in the set
num.discard(r_num)

print("Set after removing number:-", num)


# Prints the final set and its length using len().
print("\nFinal Results")
print("Final Set:-", num)
print("Total items:-", len(num))
```

```python
'''
Q9. (Practical Application)

Write a program to remove duplicate marks from a list of student scores using a set: 
scores = [85, 92, 78, 92, 85, 88, 95, 78] 
Convert the list to a set. 
Print the unique scores. 
Convert the set back to a sorted list and print it. 
Print the total number of unique scores.
'''

scores = [85, 92, 78, 92, 85, 88, 95, 78] 

# Converting the list to a set. 
s = set(scores)

# Printing the unique scores. 
print("\nUnique scores:-\n" ,s)

# Converting the set back to a sorted list and print it. 
l = list(s)
l.sort()
print("\nAfter Converting the set back to a sorted list and printing it:-\n" ,l)

# Printing the total number of unique scores.
print("\nPrinting the total number of unique scores:-" ,len(l))
```

```python
'''
Q10. (Mini Project - Combined Concepts)

Create a Unique Item Collector program using sets: 
Repeatedly ask the user to enter items (e.g., favorite fruits, colors, or numbers). 
Store all entered items in a set (duplicates should be ignored). 
Provide a menu with options: 
1. Add item 
2. Remove item (use discard) 
3. Show all unique items 
4. Check if an item exists 
5. Clear all items 
6. Exit 
Use a while loop for the menu and appropriate set methods. 
Add meaningful comments.
'''

# Defining an empty set to collect unique items
item_collector = set()

print("Welcome to the Unique Item Collector")

# Use a while loop for the menu and appropriate set methods. 
while True:
    print("\nMain Menu")
    print("1. Add item")
    print("2. Remove item")
    print("3. Show all unique items")
    print("4. Check if an item exists")
    print("5. Clear all items")
    print("6. Exit\n")
    
    
    # Taking menu choice input from the user
    choice = input("Enter your choice (1-6):- ")
    

    if choice == "1":
        new_item = input("\nEnter the item name to add:- ").upper()
        item_collector.add(new_item)
        print(f"'{new_item}' has been addad.")
        
    elif choice == "2":
        remove_item = input("\nEnter the item name to remove:- ").upper()
        # .discard() is used so it doesn't crash if the item is missing
        item_collector.discard(remove_item)
        print(f"'{remove_item}' has been removed (if it existed).")
     
    elif choice == "3":
        print("\nYour Unique Items Collection")
        if len(item_collector) == 0:
            print("The collection is currently empty!")
        else:
            print(item_collector)
       
    elif choice == "4":
        search_item = input("\nEnter the item name to search:- ").upper()
        if search_item in item_collector:
            print(f"Yes! '{search_item}' is present in your collection.")
        else:
            print(f"No! '{search_item}' is NOT present in your collection.")
            
    elif choice == "5":
        item_collector.clear()
        print("\nAll items cleared! Your collection is now empty.")
  
    elif choice == "6":
        print("\nThank you for using the Unique Item Collector. Goodbye!")
        break  # Terminates the while loop completely
        
    # Handling invalid menu selections safely
    else:
        print("\nInvalid choice! Please select a valid number from 1 to 6.")

```
